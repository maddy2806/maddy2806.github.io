## Part 1.4: Analysis
In this section, I will analyze survey data to determine whether optional pre-lecture videos would improve teh COMP110 course. 

```python
from data_utils import(
    read_csv_rows, 
    columnar, 
    head, 
    select, 
    concat, 
    count, 
    convert_columns_to_int, 
    average_rating,
)
import seaborn as sns 
import matplotlib.pyplot as plt 

sns.set_theme()

DATA_FILE_DIRECTORY: str = "data/"
DATA_FILE_NAME_IZZI: str = f"{DATA_FILE_DIRECTORY}survey_izzi.csv"
DATA_FILE_NAME_ALYSSA: str = f"{DATA_FILE_DIRECTORY}survey_alyssa.csv"

izzi_rows: list[dict[str, str]] = read_csv_rows(DATA_FILE_NAME_IZZI)
alyssa_rows: list[dict[str, str]] = read_csv_rows(DATA_FILE_NAME_ALYSSA)

izzi_cols: dict[str, list[str]] = columnar(izzi_rows)
alyssa_cols: dict[str, list[str]] = columnar(alyssa_rows)

combined_data: dict[str, list[str]] = concat(izzi_cols, alyssa_cols)
```

I will preview the first five rows of the combined dataset using the `head` function.

```python
head(combined_data, 5)
```

Next, I will select the columns that are most relevant to my analysis: pre-lecture videos, difficulty, pace, understanding, and prior programming experience.

```python
selected_data: dict[str, list[str]] = select(
    combined_data,
    ["pre_lecture_videos", "difficulty", "pace", "understanding", "prior_exp"],
)

head(selected_data, 5)
```

I will use the `count` function to see how many students selected each rating for pre-lecture videos.

```python
video_counts: dict[str, int] = count(selected_data["pre_lecture_videos"])
video_counts
```

I will use my helper function, `average_rating`, to calculate the average rating for pre-lecture video support.

```python
average_pre_video_support: float = average_rating(selected_data["pre_lecture_videos"])
average_pre_video_support
```

Before graphing, I will convert numeric survey columns from strings into integers.

```python
numeric_data: dict[str, list[str | int]] = convert_columns_to_int(
    selected_data,
    ["pre_lecture_videos", "difficulty", "pace", "understanding"],
)
```

This first graph shows the overall distribution of student support for optional pre-lecture videos.

```python
sns.catplot(
    data=numeric_data,
    x="pre_lecture_videos",
    kind="count",
)

plt.title("Student Support for Optional Pre-Lecture Videos")
plt.xlabel("Rating: 1 = Strongly Disagree, 7 = Strongly Agree")
plt.ylabel("Number of Students")
plt.show()
```

This graph compares course difficulty with support for pre-lecture videos. If students who find the course more difficult also support videos, that would support my idea.

```python
sns.catplot(
    data=numeric_data,
    x="difficulty",
    y="pre_lecture_videos",
    kind="box",
)

plt.title("Support for Pre-Lecture Videos by Course Difficulty")
plt.xlabel("Difficulty Rating")
plt.ylabel("Pre-Lecture Video Support")
plt.show()
```

This graph compares course pace with support for pre-lecture videos. This helps determine whether students who feel the course moves quickly also want preview materials.

```python
sns.relplot(
    data=numeric_data,
    x="pace",
    y="pre_lecture_videos",
    kind="scatter",
)

plt.title("Course Pace vs. Support for Pre-Lecture Videos")
plt.xlabel("Pace Rating")
plt.ylabel("Pre-Lecture Video Support")
plt.show()
```

This final graph compares prior programming experience with support for pre-lecture videos. This helps determine whether students with different backgrounds respond differently to the idea.

```python
sns.catplot(
    data=numeric_data,
    x="prior_exp",
    y="pre_lecture_videos",
    kind="box",
)

plt.title("Support for Pre-Lecture Videos by Prior Experience")
plt.xlabel("Prior Programming Experience")
plt.ylabel("Pre-Lecture Video Support")
plt.xticks(rotation=45)
plt.show()
```

## Part 1.5: Conclusion

My analysis suggests that optional pre-lecture videos could create value for students in COMP110. The distribution of responses shows the overall level of student support for pre-lecture videos, and the comparisons with difficulty, pace, and prior experience help show which students may benefit most.

However, this analysis does not prove that pre-lecture videos would directly improve grades or assignment performance. The data is based on student survey responses, so it shows opinions and relationships between responses rather than direct learning outcomes. A possible downside is that creating videos would require additional time from instructional staff, and students might feel overwhelmed if the videos seem like extra required work.

A good refinement would be to keep the videos short, optional, and focused on the most difficult topics. Future work could collect data on whether students actually watch the videos and whether video use is related to quiz scores, programming assignment performance, or lower office hour demand.
